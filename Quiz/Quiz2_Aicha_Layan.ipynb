Topic Detection

```python
import assemblyai as aai

audio_url = "https://github.com/AssemblyAI-Examples/audio-examples/raw/main/20230607_me_canadian_wildfires.mp3"

config = aai.TranscriptionConfig(iab_categories=True)

transcript = aai.Transcriber().transcribe(audio_url, config)

# Get the parts of the transcript that were tagged with topics
for result in transcript.iab_categories.results:
    print(result.text)
    print(f"Timestamp: {result.timestamp.start} - {result.timestamp.end}")
    for label in result.labels:
        print(f"{label.label} ({label.relevance})")

# Get a summary of all topics in the transcript
for topic, relevance in transcript.iab_categories.summary.items():
    print(f"Audio is {relevance * 100}% relevant to {topic}")
```

```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel, pipeline

def generate_text_with_gpt2(prompt, max_length=100, summarization_max_length=50):
    # Load the pre-trained GPT-2 model and tokenizer
    model_name = "gpt2"
    tokenizer = GPT2Tokenizer.from_pretrained(model_name)
    model = GPT2LMHeadModel.from_pretrained(model_name)

    # Encode the prompt and generate text using GPT-2
    input_ids = tokenizer.encode(prompt, return_tensors="pt")
    output_ids = model.generate(input_ids, max_length=max_length, num_return_sequences=1)
    output_text = tokenizer.decode(output_ids[0], skip_special_tokens=True)

    # Use summarization pipeline to summarize the generated text
    summarizer = pipeline("summarization")
    summarized_text = summarizer(output_text, max_length=summarization_max_length, min_length=5, length_penalty=2.0, num_beams=4)[0]['summary_text']

    return summarized_text

# Input paragraph
paragraph = """
Artificial intelligence (AI) is revolutionizing the way we live and work. It involves the development of computer systems that can perform tasks that typically require human intelligence. From natural language processing to image recognition, AI applications are diverse and impactful.

One of the key challenges in AI is ensuring ethical and responsible use. As AI systems become more sophisticated, there is a growing need for guidelines and regulations to govern their deployment. Privacy concerns, bias in algorithms, and the potential for job displacement are important considerations in the ethical use of AI.

Despite challenges, AI holds great promise for solving complex problems and improving efficiency across various industries. From healthcare to finance, AI applications are helping make better decisions and streamline processes. As the field continues to advance, it is crucial to balance innovation with ethical considerations to ensure a positive and inclusive impact on society.
"""

# Generate and summarize text
generated_text = generate_text_with_gpt2(paragraph)

print("Input Paragraph:")
print(paragraph)
print("\nGenerated and Summarized Text:")
print(generated_text)
```

In this task we used the GPT-2 model to summarize a paragraph that is given from the user. It will generate a summarized text based on an input paragraph. It then employs a summarization model to provide a concise summary of the generated text. The goal is to showcase how GPT-2 can be used for both text generation and summarization, particularly when given a paragraph as input.
